```python
import pickle
```

```python

with open('results_machop.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
resultkeys = results.keys()
```

```python
ResultsDict = {}
mod_num = 0
for i in resultkeys:
    mod_num += 1
    mod_dict = {}
    for j in i:
        mod_dict[j[0]] = j[1]
    for k in [1,5,10,20]:
        for l in ['precision', 'recall', 'f1-score', 'support']:
            mod_dict[f'{k} {l}'] = results[i][k]['test_report']['1'][l]
    ResultsDict[f'Model {mod_num}'] = mod_dict
```

```python
    mod_dict['1 Prec'] = results[i][1]['test_report']['1']['precision']
    mod_dict['1 Recall'] = results[i][1]['test_report']['1']['recall']
    mod_dict['1 F1'] = results[i][1]['test_report']['1']['f1-score']
```

```python
import pandas as pd
```

```python
resDF = pd.DataFrame(ResultsDict) # convert into pandas dataframe
```

```python
resDFT = resDF.transpose()
```

```python
print(f'Max f1 for 20 iters {max(resDFT["20 f1-score"])}')
print(f'Max f1 for 10 iters {max(resDFT["10 f1-score"])}')
print(f'Max f1 for 5 iters {max(resDFT["5 f1-score"])}')
print(f'Max f1 for 1 iters {max(resDFT["1 f1-score"])}')
```

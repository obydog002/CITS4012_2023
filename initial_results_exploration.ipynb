```python
import pickle
```

```python

with open('results.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
resultkeys = results.keys()
```

```python
ResultsDict = {}
mod_num = 0
for i in resultkeys:
    mod_num += 1
    mod_dict = {}
    for j in i:
        mod_dict[j[0]] = j[1]
    for k in [1,5,10,20]:
        for l in ['precision', 'recall', 'f1-score', 'support']:
            mod_dict[f'{k} {l}'] = results[i][k]['test_report']['1'][l]
            mod_dict[f'{k} training {l}'] = results[i][k]['train_report']['1'][l]
    ResultsDict[f'Model {mod_num}'] = mod_dict
```

```python
import pandas as pd
```

```python
resDF = pd.DataFrame(ResultsDict) # convert into pandas dataframe
```

```python
resDFTNDC = resDF.transpose()
```
